## Broj nastupa na kupu

```python
#%matplotlib inline
%matplotlib notebook
import matplotlib as mpl
import matplotlib.pyplot as pp
#import numpy as np

x = [2001, 2002, 2003]
y = [34, 19, 22]

pp.plot(x, y)
pp.xticks(x)
pp.ylim([0,40])

pp.xlabel('godina')
pp.ylabel('Broj timova na kupu')
pp.title('Povijest Kupa Hrvatske')
```
