## MDF reading and plotting

How to read and plot MDF data using Python package [mdfreader](https://pypi.python.org/pypi/mdfreader):

```python
import matplotlib.pyplot
matplotlib.pyplot.rcParams["figure.figsize"] = (20,8) # default matplot lib figure size

import mdfreader
f = mdfreader.mdf('afs_recorder.mdf', channelList=['M', 'N']) # load only some channels from mdf file
#f.keys() # channels names
f.plot({'N','M'})

```
